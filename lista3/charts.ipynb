```python
from matplotlib import pyplot as plt 
import os
```

```python
# (n, m, minW, maxW, t)
def extractData(directory):
    data = []

    for filename in os.listdir(directory):
        filepath = os.path.join(directory, filename)
        with open(filepath, 'r') as file:
            lines = file.readlines()
            infoLine = lines[2].strip()
            timeLine = lines[3].strip()
            info = [int(x) for x in infoLine.split()[1:]]
            data.append((*info, float(timeLine.split()[1])))

    return data
```

```python
def generateLongNChart(directory):
    dijkstra = extractData(directory + '/dijkstra')
    dial = extractData(directory + '/dial')
    radix = extractData(directory + '/radix')

    x = [val[0] for val in dijkstra]
    yDijkstra = [val[4] for val in dijkstra]
    yDial = [val[4] for val in dial]
    yRadix = [val[4] for val in radix]

    plt.scatter(x, yDijkstra, label = 'dijkstra')
    plt.scatter(x, yDial, label = 'dial')
    plt.scatter(x, yRadix, label = 'radix')

    plt.title('Dijkstra Variants Comparision\nLong-n Family; n = C, m = 4n')
    plt.xlabel('n')
    plt.ylabel('time [ms]')
    plt.xscale('log')

    plt.legend()
    plt.show()
```

```python
generateLongNChart('tests/outputs/Long-n')
```
